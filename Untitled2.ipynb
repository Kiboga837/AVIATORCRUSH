```python
import streamlit as st
import numpy as np
import hashlib
import uuid
import time

st.set_page_config(page_title="Aviator Crash Game", layout="centered")

def generate_crash_point():
    seed = str(uuid.uuid4())
    hash_value = hashlib.sha256(seed.encode()).hexdigest()
    number = int(hash_value[:8], 16)
    return round(max(1.0, (2**32 / (number + 1))), 2)

st.title("‚úàÔ∏è Aviator Crash Game")

if "crash" not in st.session_state:
    st.session_state.crash = generate_crash_point()
    st.session_state.multiplier = 1.0
    st.session_state.running = False
    st.session_state.result = ""

if st.button("START ROUND"):
    st.session_state.crash = generate_crash_point()
    st.session_state.multiplier = 1.0
    st.session_state.running = True
    st.session_state.result = ""

if st.button("CASH OUT"):
    if st.session_state.running:
        st.session_state.running = False
        st.session_state.result = f"‚úÖ Cashed out at {st.session_state.multiplier:.2f}x"

placeholder = st.empty()

if st.session_state.running:
    while st.session_state.multiplier < st.session_state.crash:
        st.session_state.multiplier += 0.02
        placeholder.metric("Multiplier", f"{st.session_state.multiplier:.2f}x")
        time.sleep(0.1)

    st.session_state.running = False
    st.session_state.result = f"üí• CRASHED at {st.session_state.crash}x"

if st.session_state.result:
    st.info(st.session_state.result)
```
